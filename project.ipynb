```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv('Untitled spreadsheet - T1.csv')
```

```python
dataset
```

```python
dataset.drop((['Unnamed: 5']),axis=1,inplace=True)
dataset
```

```python
dataset.drop((['Date/Time']),axis=1,inplace=True)
dataset
```

```python
dataset.isnull().any()
```

```python
x=dataset.iloc[:,0:3].values
x
```

```python
y=dataset.iloc[:,3:4].values
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
x_test
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
mr = LinearRegression()
```

```python
mr.fit(x_train,y_train)
```

```python
y_predict =mr.predict(x_test)
```

```python
y_predict
```

```python
mr.predict([[380.047790,5.311336,259.994904]])
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_predict)
```
